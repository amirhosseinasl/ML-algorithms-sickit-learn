```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

**One Hot Encoder**

```python
df = pd.read_csv('house-prices.csv')
new = df[['Brick' , 'Neighborhood']]
```

```python
enc = OneHotEncoder()
tra_array = enc.fit_transform(new)
#enc.categories_

transformed = pd.DataFrame(tra_array.toarray() , columns = enc.get_feature_names_out())
transformed
```

```python
house_prices = pd.concat([df , transformed] , axis = 1)
house_prices = house_prices.drop(['Brick' , 'Neighborhood'] , axis = 1)
house_prices
```
