```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import mean_squared_error
```

**Linear Regression**

```python
house_prices = pd.read_csv('house-prices.csv')
target = house_prices['Price']
features = house_prices[['SqFt' , 'Bedrooms' , 'Bathrooms' , 'Offers']]
X_train , X_test , y_train , y_test = tts(features , target , test_size = 0.33 , random_state = 42 , shuffle = False)
house_prices
```

```python
linreg_model = LinearRegression()
linreg_model.fit(X_train , y_train)
```

```python
y_pred = linreg_model.predict(X_test)
```

```python
y_pred
```

```python
mse = mean_squared_error(y_pred , y_test)
np.sqrt(mse).round(3)
```

```python
linreg_model.coef_
```

```python
np.abs(linreg_model.intercept_)
```
