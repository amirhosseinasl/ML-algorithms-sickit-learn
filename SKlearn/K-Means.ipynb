```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sbn
```

**K-Means**

```python
df = pd.read_csv('Mall_Customers.csv')
km = KMeans(n_clusters = 5)

km.fit(df[['Annual Income (k$)' , 'Spending Score (1-100)']])
df['label'] = km.labels_
```

```python
km.labels_
```

```python
sbn.scatterplot(x = df['Annual Income (k$)'] , y =  df['Spending Score (1-100)'] , hue = df['label'] , palette = 'summer')
```
