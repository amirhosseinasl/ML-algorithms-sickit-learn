```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split as tts
from sklearn.tree import plot_tree
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

**DecisionTreeClassifier**

```python
x , y = load_breast_cancer(return_X_y = True , as_frame = True)
```

```python
X_train , X_test , y_train , y_test = tts(x , y , test_size = 0.33 , shuffle = False)
```

```python
dtc = DecisionTreeClassifier(max_depth = 6 , class_weight = {0 : 0.3 , 1 : 0.7})
dtc.fit(X_train , y_train)
```

```python
plot_tree(dtc , filled = True , fontsize = 6 , feature_names = x.columns)
```

```python
y_pred = dtc.predict(X_test)
#y_pred = dtc.predict_proba(X_test)
metrics.precision_score(y_pred , y_test)
```

```python
metrics.recall_score(y_pred , y_test)
```

```python
metrics.confusion_matrix(y_pred , y_test)
```
